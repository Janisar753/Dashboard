```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\91963\Downloads\March sales.csv")
```

```python
df
```

```python
t=df[df.partners=='Aegis'].pivot_table(index='seller_id', columns='service_profile', values='units',aggfunc=sum)
t
```

```python
fbf=t.agg({'FBF':["sum"],'FBF_LITE':["sum"],'NON_FBF':["sum"]})
```

```python
fbf
```

```python
fbf['Total']=fbf['FBF']+fbf['FBF_LITE']+fbf['NON_FBF']
```

```python
fbf['FBF']/fbf['Total']*100  #this is FBF%
```

```python
s=t.dropna(subset=['FBF'])
s['Total']=s['FBF']+s.FBF_LITE.fillna(0)+s.NON_FBF.fillna(0)
s
```

```python
adop=s.agg({'FBF':["sum"],'NON_FBF':["sum"]})
adop
```

```python
adop['total']=adop['FBF']+adop['NON_FBF']
adop
```

```python
adop_total=adop['FBF']/adop['total']*100
```

```python
adop_total  #this is adoption FBF%
```
